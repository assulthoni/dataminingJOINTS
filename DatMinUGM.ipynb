```
import pandas as pd
import numpy as np
import csv
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB
```

Reading data Train and test

```
dataTrainUrl = 'https://raw.githubusercontent.com/assulthoni/dataminingJOINTS/master/train.csv' 
dataTrain = pd.read_csv(dataTrainUrl)

dataTestUrl = 'https://raw.githubusercontent.com/assulthoni/dataminingJOINTS/master/test_data.csv' 
dataTest = pd.read_csv(dataTestUrl)

dataTrain = dataTrain.drop('id',axis=1) 
dataTest = dataTest.drop('id',axis=1)

dataTrain.head()
```

Cleaning Data NaN

```
# clearDataTrain = dataTrain.dropna()
# clearDataTest = dataTest.dropna()

clearDataTrain = dataTrain.fillna(0)
clearDataTest = dataTest.fillna(0)
```

```
le = preprocessing.LabelEncoder()

k1 = clearDataTrain.fac_1
k2 = clearDataTrain.fac_2
k3 = clearDataTrain.fac_3
k4 = clearDataTrain.fac_4
k5 = clearDataTrain.fac_5
k6 = clearDataTrain.fac_6
k7 = clearDataTrain.fac_7
k8 = clearDataTrain.fac_8
k9 = clearDataTrain.poi_1
k10 = clearDataTrain.poi_2
k11 = clearDataTrain.poi_3
k12 = clearDataTrain['size']
k13 = clearDataTrain.price_monthly
k14 = clearDataTrain.room_count
k15 = clearDataTrain.total_call
kelas = le.fit_transform(clearDataTrain.gender)
```

```
# var1 = varRange.as_matrix()
features = np.column_stack((k1,k2,k3,k4,k5,k6,k7,k8,k9,k10,k11,k12,k13,k14,k15))
model = GaussianNB()
model.fit(features,kelas)

dataPredict = clearDataTest
predicted = model.predict(dataPredict)

a = np.array(predicted)
a = a.astype(object)
# np.savetxt("D:\kuliah\LOMBA\Data Mining UGM\submitted.csv", a,fmt='%s', delimiter=",")
print (predicted)
```
